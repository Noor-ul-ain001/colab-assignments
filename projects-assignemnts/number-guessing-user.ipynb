```python
import random  #package use for generating a random number

print("ðŸ™‚Welcome to my Number Guessing game")
print("Kindly choose a number between 50 - 100 \n there are only 5 attemptsðŸ™ƒ")
chance = 5 #define how many chances
number_to_guess= random.randrange(50, 100)  #define specific range of a number
guess_counter = 0 #initial value of count

while guess_counter < chance:
    guess_counter += 1  #on each attempt chance to select decreases
    my_guess = int(input("\nplease select a number: "))   #insert number

    if (my_guess == number_to_guess ):
        print(f"ðŸ¥³Congrats you have selected right number {number_to_guess} in {guess_counter} attempts")
        break
    elif (my_guess != number_to_guess and guess_counter >= chance ):
        print("ðŸ’—incorrect guess. better luck next time")
    elif (my_guess > number_to_guess):
        print("ðŸ”ºyour guess is too high please select low value...")
    elif (my_guess < number_to_guess):
        print("ðŸ”»your guess is too low please select high value...")

```
