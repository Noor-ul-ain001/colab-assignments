***rock-paper-scissor***

```python

import random

user_choice : str = input("enter your choice,  rock,  paper ,scissor:  ").lower()
choices : list[str] = ["rock","paper", "scissor"]

computer_choice: str = random.choice(choices)
# print(user_choice, computer_choice)

if user_choice not in choices:
    print("Invalid choice! Please enter  rock, paper, or scissor.")

elif (user_choice == computer_choice):
    print(f"user choose: {user_choice} and computer choose: {computer_choice}")
    print("It's a tie")

elif (user_choice == "rock" and computer_choice == "scissor"):
    print(f"user choose: {user_choice} and computer choose: {computer_choice}")
    print("you win")

elif (user_choice == "scissor" and computer_choice == "paper"):
    print(f"user choose: {user_choice} and computer choose: {computer_choice}")
    print("you win")

elif (user_choice == "paper" and computer_choice == "rock"):
    print(f"user choose: {user_choice} and computer choose: {computer_choice}")

    print("you win")

else:
    print(f"user choose: {user_choice} and computer choose: {computer_choice}")

    print("Computer win")

```
