# ***hangman-python-project***

```python
import random

words = [
  "keyboard", "adventure", "pineapple", "champion", "sunflower",
  "butterfly", "mystery", "astronaut", "lightning", "treasure",
  "elephant", "galaxy", "horizon", "volcano", "whisper",
  "jellyfish", "carousel", "zeppelin", "fireworks", "pumpkin",
  "chocolate", "snowflake", "crocodile", "umbrella", "paradise",
  "tornado", "lantern", "moonlight", "dragonfly", "waterfall"
]

word = random.choice(words)  # Choose a random word
guessed_letters = []
attempts = 6

print("Welcome to the Hangman Game!")
print("_ " * len(word))  # Print underscores for hidden letters

while attempts > 0:
    guess = input("Guess a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single valid letter.")
        continue

    if guess in guessed_letters:
        print("This letter is already guessed. Choose another.")
        continue

    guessed_letters.append(guess)

    if guess in word:
        print("Correct guess!")
    else:
        attempts -= 1
        print(f"Wrong guess! {attempts} attempts left.")

    displayed_word = " ".join([letter if letter in guessed_letters else "_" for letter in word])
    print(displayed_word)

    if "_" not in displayed_word:
        print(f"Congratulations! The correct word is '{word}'.")
        break  # End game if the word is fully guessed

else:
    print(f"Game over! The correct word was '{word}'.")
```
