# ***count-down-timer***

```python
import time
user_input = int(input("enter your time in seconds: "))

for i in range(user_input, 0, -1):
    second = i % 60
    minutes = int(i / 60) % 60
    hours = int(i /3600)
    time.sleep(1)
    print(f"{hours:02}:{minutes:02}:{second:02}")
    # print(i)

print("time's up")
```
